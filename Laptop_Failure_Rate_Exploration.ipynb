# Investigating Rates of Failure Reports of Laptops in Amazon Reviews

[FaultFindr](https://faultfindr.xyz) is a web-app that I created which identifies sentences within Amazon reviews that are discussing Laptop problems that people experienced. With this model working, I would now like to investigate the patterns in the rates of failure for laptops across various laptop properties such as brand, mean rating, etc. That data exploration is the purpose of this document.

There are a few caveats in this analysis that I would like to state upfront:
- the model that identifies failures within the amazon reviews are not necessarily unbiased with respect to the variables that I am exploring (for example brand, screen size, ...). The precision and recall of this model may depend on these parameters, and hence the variation of the failure rates with respect to these parameters could be at least partially attributable to bias in the model.
- the estimated failure rates are not estimates of failure rates in the general population of laptops sold, but rather the fraction of the reviews that contain discussion of laptop failure. It can then be a little dangerous to extrapolate from one to the other since:
    - if people have a problem with their laptops they are probably more likely to write a review than if they did not have a problem which increases the rate of failure in reviews with respect to rate of failure in the general population.
    - people who tend to buy a certain type of laptop may be more likely to write reviews about laptop failure than people who tend to buy other types of laptops, which skews the rates of failure amongst laptops.
- the model has a recall of about 50% (I optimized the precision of the model with fixed recall).

## Correlation between mean rating and rate of failure

Naively, I might think that I could identify the best laptops by looking at the mean rating on Amazon provided that there are enough reviews. However, a quick analysis shows that the failure rate observed within the reviews is not very correlated with the mean rating. 

Here, I am only using laptops that have more than 50 reviews (to try to reduce the uncertainties in the failure rates without removing too much data), and I am approximating the uncertainty in the rates $r = n_{\rm failure}/n_{\rm total}$ by the approximate formula $r = \sqrt{n_{\rm failure}}/n_{\rm total}$, which is reasonably valid when $n_{\rm failure}\gg1$.

Assuming a simple linear relationship between the rate of failure in reviews and the mean rating, and renormalizing the fit coefficients that result by the measured $\chi^2\approx 2.4$, we find that the slope is consistent with zero within $1\sigma$.

This suggests that the failure rate in the reviews is independent of the mean product rating, and thus provides a possibly more useful metric for evaluating products than the mean star rating if you are trying to minimize the likelihood that you will end up with a defunct product.

```python
# only use products with a sufficiently large number of reviews
sub = l[l.n_total > 50]

# perform a linear regression:
regr = PolynomialFit1D(sub.mean_rating, sub.rate, sub.drate, order=1,
                       renormalize_error=True,x_name='s', y_name='r')

# plot the outcome:
regr.plot(legend=True)

plt.ylim([.01,1])
plt.yscale('log')
plt.xlabel('$s$ (mean rating)')
_ = plt.ylabel('$r$ (rate of failure)')
```

## Variation in Failure Rates Across Brands

Grouping the available laptops by brand, we can look for variation in the laptop failure rates across brands. The results are not incredibly surprising - Apple products clearly have the lowest product failure rates, and some set of PC brands such as Dell and HP have similar but higher failure rates. Among the PC brands, Lenovo and Samsung seem to have the lowest product failure rates. Perhaps the biggest surpise here is that Acer, which for the most part makes low end chromebooks, has a pretty low failure rate, just above Apple's. The failure rates vary by about a factor of about $2\times$ between $8\%-15\%$. However, keep in mind all of the possible caveats to the interpretation of this data that I outlined in the introduction to this document.

```python
brand_rates = pd.DataFrame(
                [(brand, 
                l[l.brand == brand].n_failure.sum(),
                l[l.brand == brand].n_total.sum()) for brand in l.brand.unique()],
              columns = ['brand', 'n_failure', 'n_total'])
brand_rates['rate'] = np.array(brand_rates['n_failure'])/brand_rates['n_total']
brand_rates['drate'] = np.sqrt(brand_rates['n_failure'])/brand_rates['n_total']

data = brand_rates[brand_rates.n_total > 250].sort_values(by='rate')

sns.barplot(data.brand, 2 * data.drate, bottom=data.rate - data.drate, alpha=.5)

ylim = plt.gca().get_ylim()
plt.ylim([0, ylim[1]])
plt.xticks(rotation=30)
plt.ylabel('rate')
plt.title('rates of failure in reviews across laptop brands')
```

## Should you buy refurbished?

We might expect refurbished computers to be more likely to fail than new computers since they tend to have older components and they have been modified by a process that probably hasn't been as refined as the main manufacturing process. Looking at the data, there doesn't seem to be enough data to come to any conclusions. The failure rates for refurbished and new laptops are not statistically different within the resolution of about $5\%$.

```python
refurbished_rates = pd.DataFrame(
                [(refurbished, 
                l[l.refurbished == refurbished].n_failure.sum(),
                l[l.refurbished == refurbished].n_total.sum()) for refurbished in l.refurbished.unique()],
              columns = ['refurbished', 'n_failure', 'n_total'])
refurbished_rates['rate'] = np.array(refurbished_rates['n_failure'])/refurbished_rates['n_total']
refurbished_rates['drate'] = np.sqrt(refurbished_rates['n_failure'])/refurbished_rates['n_total']

refurbished_rates
```

## Do larger laptops have more problems?

We might expect that if we buy a laptop with a very large screen, that the screen is at a larger risk of breaking. I decided to look at the variation in the failure rates as a function of screen size, and was surprised to see an even larger effect than I expected - about 1% increase in failure rate per inch of screen (presumably measured along the diagonal).

```python
# keep only those entries where I was able to extract a screen size
sub = l[np.logical_not(np.isnan(l.screen_size))]

# at least one of the entries has an absurdly large value - truncate that
sub = sub[sub.screen_size < 100]

# group by the discrete screen sizes that are sold
screen_size_grouped = sub.groupby('screen_size')

# evaluate the number of failures and rates in each group
n_failure = screen_size_grouped.n_failure.sum()
n_total   = screen_size_grouped.n_total.sum()
screen_size_rates = pd.DataFrame([n_failure, n_total]).T
screen_size_rates['rate'] = screen_size_rates['n_failure']/screen_size_rates['n_total']
screen_size_rates['drate'] = np.sqrt(screen_size_rates['n_failure'])/screen_size_rates['n_total']

# only look at groups that have at least 50 reviews so that my rate
# uncertainty estimates are reasonable
screen_size_rates = screen_size_rates[screen_size_rates.n_total > 50]

# perform linear regression
regr0 = PolynomialFit1D(screen_size_rates.index, 
                       screen_size_rates.rate, 
                       screen_size_rates.drate, 
                       order=1,
                       renormalize_error=True,
                       x_name='s', y_name='r')

# plot the results
regr0.plot(legend=True)

plt.ylim([.01,1])
plt.yscale('log')
plt.xlabel('screen size $s$ (inches)')
_ = plt.ylabel('rate of failure $r$')
```

## Variation in Screen Quality Across Brands

Since this effect was so large, I decided to look a little closer at the effect, to see if the variation in failure rate with increasing screen varies across manufacturers -- i.e. do some manufacturers make better screens? For the most part, it looks like this effect varies too significantly across brands, though the differences are large enough that is is reasonable to say, for example, that Apple screens seem to be above average in quality, and Toshiba tend to use screens that are below average in quality.

```python
# keep only those entries where I was able to extract a screen size
sub = l[np.logical_not(np.isnan(l.screen_size))]

# at least one of the entries has an absurdly large value - truncate that
sub = sub[sub.screen_size < 100]

# group by the discrete screen sizes that are sold and by brand
screen_size_grouped = sub.groupby(['screen_size', 'brand'])

# evaluate the number of failures and rates in each group
n_failure = screen_size_grouped.n_failure.sum()
n_total   = screen_size_grouped.n_total.sum()
screen_size_rates = pd.DataFrame([n_failure, n_total]).T
screen_size_rates['rate'] = screen_size_rates['n_failure']/screen_size_rates['n_total']
screen_size_rates['drate'] = np.sqrt(screen_size_rates['n_failure'])/screen_size_rates['n_total']

# only look at groups that have at least 25 reviews so that my rate
# uncertainty estimates are reasonable
screen_size_rates = screen_size_rates[screen_size_rates.n_total > 25]
screen_size_rates.reset_index(inplace=True)

# evaluate the slope of failure rate vs screen size for each brand:
slopes = []
regrs  = []
for brand in screen_size_rates.brand.unique():
    subset = screen_size_rates[screen_size_rates.brand == brand]
    x = subset.screen_size
    y = subset.rate
    dy= subset.drate
    if len(x) > 2:
        regr = PolynomialFit1D(x, y, dy, 
                               order=1,
                               renormalize_error=True,
                               x_name='s', y_name='r',
                               label=brand)
        
        slope = regr.get_coefficients().iloc[1]
        slopes.append((brand, slope['value'], slope['dvalue']))
        
        regrs.append(regr)

# aggregate the resulting slopes in a dataframe
slopes = pd.DataFrame(slopes, columns=['brand', 'slope', 'dslope'])
slopes.sort_values(by='slope', inplace=True)
slopes.reset_index(drop=True, inplace=True)

# plot the results
sns.barplot(slopes.brand, 
            (2 * slopes.dslope), 
            bottom=(slopes.slope - slopes.dslope), alpha=.5)

plt.xticks(rotation=30)
plt.ylabel('(rate of failure) / (inch of screen)')
plt.title('increase in failure rate with increasing screen size\nacross laptop brands')

xlim = plt.gca().get_xlim()
plt.plot(xlim, [0, 0], '--k')
x, dx = regr0.coefficients[1], regr0.dcoefficients[1]
plt.plot(xlim, [x-dx, x-dx], '-k')
plt.plot(xlim, [x+dx, x+dx], '-k')
```
